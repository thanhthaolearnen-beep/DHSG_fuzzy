```python
import numpy as np
import matplotlib.pyplot as plt

def water_quality(pH):
    if pH <= 5:
        return 0
    elif 5 < pH < 6:
        return (pH - 5) / 1
    elif 6 <= pH <= 8.5:
        return 1
    elif 8.5 < pH < 9.5:
        return (9.5 - pH) / 1
    else:
        return 0

x = np.linspace(0, 14, 400)
y = [water_quality(v) for v in x]

plt.figure(figsize=(8,4))
plt.plot(x, y, linewidth=2, color='blue')
plt.title("Hàm thuộc cho chất lượng nước uống theo pH")
plt.xlabel("Giá trị pH (0–14)")
plt.ylabel("Mức độ thuộc μ(pH)")
plt.grid(True)
plt.ylim(-0.05, 1.05)
plt.show()
```
