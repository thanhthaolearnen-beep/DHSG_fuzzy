```python
%load_ext autoreload
%autoreload 2

import numpy as np, os
import matplotlib.pyplot as plt
```

```python
exp_dir = 'exp1_251127'
os.makedirs(exp_dir, exist_ok=True)
```

## Ví dụ 1

```python
X = np.array([30, 50, 75,  90 , 100, 125, 150, 200])
A = np.array([ 0,  0,  0,  0.1, 0.5, 0.8,   1,   1])
B = np.array([0.2, 0.4, 0.7, 0.9, 1.0, 0.7, 0.4, 0.1])
C = np.array([0.3, 0.9, 1.0, 1.0, 0.9, 0.5, 0.1, 0.0])

for s_var in ['X', 'A', 'B', 'C']:
    print(f'{s_var} = {globals()[s_var]}, len = {len(globals()[s_var])}')
```

```python
plt.figure(figsize=(12,6))
plt.title("ĐỒ THỊ BIỂU DIỄN HÀM THUỘC CỦA TẬP MỜ")

plt.plot(X, A, label="Tập mờ A = { μA(x) }", color="blue", marker="o")
plt.plot(X, B, label="Tập mờ B = { μB(x) }", color="red", marker="x")
plt.plot(X, C, label="Tập mờ C = { μC(x) }", color="green", marker="s")

plt.xlabel("Không gian nền X")
plt.ylabel("Mức độ phụ thuộc  μ(x)")

plt.legend(loc="best")
plt.grid(alpha=0.3)
plt.savefig(f'{exp_dir}/fig1.png')
plt.show()
```
