## Bắt đầu

Xây dựng một hàm thuộc hợp lý cho các tập mờ sau, biểu diễn tốc độ của các phương
tiện di chuyển trên đường cao tốc (giới hạn từ 0–75 mph):

* a. Nhanh  
* b. Vừa  
* c. Chậm

### Giả sử ta chọn các khoảng tốc độ như sau:

| mph|mô tả|
|:-----|-----:|
|0-50|chậm|
|50-65|trung bình|
|65-75|nhanh|

*ta có:*    
$\mu_A = \{\frac{1}{0}, \frac{1}{50}, \frac{0}{65}, \frac{0}{75} \}$    

$\mu_B = \{\frac{0}{0}, \frac{0}{50}, \frac{1}{65}, \frac{1}{75}\}$

trong đó:  
$\mu_A$ là biểu diễn **tập mờ chậm**  
$\mu_B$ là biểu diễn **tập mờ nhanh**  

### <u><i> Vẽ đồ thị</i></u>


```python
%pip install numpy matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Không gian nền
X = np.array([0,50,65,75])

# Hàm thuộc
mu_A = np.array([1,1,0,0]) # slow
mu_B = np.array([0,0,1,1]) # fast

# chart
plt.figure(figsize=(8,5))
plt.plot(X, mu_A, marker='o', label='Tập mờ A (chậm)')
plt.plot(X, mu_B, marker='o', label='Tập mờ B (nhanh)')

plt.title("Đồ thị tập mờ")
plt.xlabel("Không gian nền X = {0,50,65,75}")
plt.ylabel("Mức độ phụ thuộc μ(x)")
plt.ylim(-0.05, 1.05)
plt.legend()
plt.grid(True)
plt.show()
```

### Hàm thuộc liên tục 

Bằng cách nội suy tuyến tính qua các điểm đã cho. Ký hiệu công thức chung cho đường thằng qua hai điểm $P(x_P,y_P)$,$Q(x_Q, y_Q)$ là:  

$y(x)=y_p+(y_Q-y_P)\cdot\frac{x-x_P}{x_Q-x_P}$

**1.Tập mờ $A$ - "chậm"**

Từ 2 điểm  
* $P(0,1)$,$Q(50,1)$ ta có:  $y=1+(1-1)\cdot\frac{x-0}{50-0} = 1$
* $P(50,1)$,$Q(65,0)$ ta có: $y=1+(0-1)\cdot\frac{x-50}{65-50}=\frac{65-x}{150}$
* $P(65,0)$,$Q(75,0)$ ta có: $y=0+(0-0)\cdot\frac{x-65}{75-65}=0$

Từ đó ta có:  
$$\mu_A=\begin{cases}
1, & x \le 50,\\
\frac{65-x}{15}, & x \in[50,65),\\
0, & x \ge 65
\end{cases}
$$

**2.Tập mờ $B$ - "nhanh"**

Từ 2 điểm
* $P(0,0)$,$Q(50,0)$ ta có: $y=0$ 
* $P(50,0)$,$Q(65,1)$ ta có: $y=0+(1-0)\cdot\frac{x-50}{65-50}=\frac{x-50}{15}
* $P(65,1)$,$Q(75,1)$ ta có: $y=1$

Từ đó ta có:
$$\mu_B=\begin{cases}
0, & x \le 50,\\
\frac{x-50}{15}, & x \in[50,65),\\
1, & x \in[65,75]
\end{cases}
$$
