# Bắt đầu

**Đề bài:**  
Bài 3. Cho các hình trụ có tỷ số ( d/h ) (đường kính trên chiều cao) như trong Hình.  
Hãy xây dựng hàm thuộc cho mỗi dạng sau, và giải thích nguyên nhân của sự chồng lấn giữa
các hàm thuộc đó:  
a. Thanh trụ (Rod)  
b. Trụ (Cylinder)  
c. Đĩa (Disk)  



<u><i>Ta có:</i></u>  

Biến đầu vào: $r=\frac{d}{h}$  
Thay đổi: r từ 0->∞
Phân loại: có 3 nhóm (thanh trụ, trụ, đĩa)  

<u><i>Theo ngôn ngữ tự nhiên thì:</i></u>  

Thanh trụ: vật có đường kính nhỏ hơn chiều cao rất nhiều, thanh mảnh  
Trụ: vật có đường kính gần như bằng chiều cao, hình dạng cân đối  
Đĩa: vật có đường kính lớn hơn chiều cao rất nhiều, hình dẹt, rộng

<u><i>Bảng mô tả:</i></u>
|Nhóm|Bất đẳng thức|Suy ra|
|---|---|---|
|Thanh trụ|d<<h|$r=\frac{d}{h}$ << 1, tức r gần 0|
|Trụ|d≈h|$r=\frac{d}{h}$ ≈ 1, tức r gần 1|
|Đĩa|d>>h|$r=\frac{d}{h}$ >> 1, tức r rất lớn|  

* Thanh trụ:  
r càng nhỏ thì càng chắc chắn 100% là thanh trụ, và $\mu=1$  
Ví dụ: r = 0.1 => chắc chắn 100% hình thanh trụ tức $\mu=1$  
r = 0.5 => độ chắc chắn % thanh trụ giảm (có thể còn 70%) tức $\mu$ tiến về 0  
r = 1 => chắc chắn ko còn là hình thanh trụ tức $\mu=0$  

* Trụ:  
r = 1 thì chắc chắn 100% là hình trụ (d gần bằng h), tức $\mu=1$  
r = 0.5 => độ chắc chắn % trụ tăng(nó vừa trụ vừa thanh trụ), độ chắc chắn % thanh trụ giảm dần  tức $\mu$ < 1  
r = 1.5 => độ chắc chắn % trụ giảm(nó vừa trụ vừa đĩa), độ chắc chắn % đĩa tăng dần tức $\mu$ > 1  

* Đĩa
r = 5 thì chắc chắn 100% đĩa (d lớn hơn hẳn h) $\mu=1$

**Hàm thuộc  

Ta có phương trình đường thẳng đi qua hai điểm: $\frac{\mu-\mu_1}{\mu_2-\mu_1}=\frac{r-r_1}{r_2-r_1}$  

* Thanh trụ: hình thang giảm đi qua hai điểm a,b  

$$\mu=\begin{cases}
1, & x \le a\\
\frac{b-r}{b-a}, & x \in[a,b]\\
0, & x \ge b\\
\end{cases}
$$

* Trụ: hình tam giác  

$$\mu=\begin{cases}
0, & x \le a\\
\frac{r-a}{b-a}, & x \in[a,b]\\
\frac{c-r}{c-b}, & x \in[b,c]\\
0, & x \ge b\\
\end{cases}
$$  

* Đĩa: hình thang tăng  

$$\mu=\begin{cases}
0, & x \le a\\
\frac{r-a}{b-a}, & x \in[a,b]\\
1, & x \ge b\\
\end{cases}
$$





### Giải thích nguyên nhân sự chồng lấn

**Nguyên nhân 1**: Bản chất liên tục của tỷ số r = d/h  
Tỷ số r thay đổi liên tục từ 0 → ∞, không có điểm "nhảy bậc" nào trong tự nhiên.  

**Nguyên nhân 2**: Ranh giới ngôn ngữ vốn dĩ mơ hồ  
Các từ "thanh", "trụ", "đĩa" là thuật ngữ ngôn ngữ (linguistic terms), không phải định nghĩa toán học chính xác.  
Con người khi nhìn vật thể có r = 0.9 sẽ nói "trông hơi giống trụ nhưng cũng khá giống thanh" → Logic mờ mô phỏng đúng điều này.  

**Nguyên nhân 3**: Không có "chuyên gia đồng thuận" về ngưỡng phân loại  
Nếu hỏi 10 kỹ sư: "r = bao nhiêu thì hết là Rod, bắt đầu là Cylinder?"  

Người nói: r = 0.5  
Người nói: r = 0.7  
Người nói: r = 1.0  

→ Sự bất đồng này chính là nguồn gốc của vùng chồng lấn — vùng chồng lấn thể hiện sự không chắc chắn trong tri thức chuyên gia.  

**Nguyên nhân 4**: Mục đích kỹ thuật — tránh "vách đứng"
