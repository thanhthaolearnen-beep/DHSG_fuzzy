## Bài làm  


Giả sử tập nền X = {5000, 6000, 9000, 10000, 20000, 50000, 60000, 90000, 100000}  

Tập A có giá trị: A = {$\frac{0}{5000}, \frac{0}{6000}, \frac{0}{9000}, \frac{0}{10000}, \frac{0.3}{20000}, \frac{1}{50000}, \frac{0.8}{60000},\frac{0.2}{90000}, \frac{0}{100000}$}      

Tập B có giá trị: B = {$\frac{0}{5000}, \frac{0}{6000}, \frac{0.8}{9000}, \frac{1}{10000}, \frac{0.3}{20000}, \frac{0}{50000}, \frac{0}{60000},\frac{0}{90000}, \frac{0}{100000}$}    

Phép toán trên không gian nền:  

|Câu|Công thức|  
|---|---|  
|$\tilde{A} \cup \tilde{B}$| max($\mu_A, \mu_B$)|  
|$\tilde{A} \cap \tilde{B}$| min($\mu_A, \mu_B$)|  
|$\bar{A}, \bar{B}$|1-$\mu_A$, 1-\$mu_B$|  
|$\tilde{A}-\tilde{B}$| min($\mu_A,1-\mu_B$)|  
|De Morgan| min{1-$\mu_A, 1-\mu_B$}|  

Từ đó ta có bảng giá trị: 

|X|$\tilde{A}$|$\tilde{B}$|$\bar{A}$|$\bar{B}$|$\tilde{A}\cup\tilde{B}$|$\tilde{A}\cap\tilde{B}$|$\tilde{A} - \tilde{B}$|$\widetilde{A \cup B} = \bar{A} \cap \bar{B}$ De Morgan|
|---|---|---|---|---|---|---|---|---|  
|5000| 0| 0| 0|0 |0 | 0| 0|0 |  
|6000| 0| 0| 0|0 |0 | 0| 0|0 |  
|9000| 0.8| 0| 0.2|0 |0.8 | 0| 0|0 |  
|10000| 1| 0| 0|0 |1 | 0| 0|0 |  
|20000| 0.3| 0.3| 0.7|0.7 |0.3 | 0.3| 0.3|0.7 |  
|50000| 0| 1| 1|0 |1 | 0| 0|0 |   
|60000| 0| 0.8| 1|0.2 |0.8 | 0| 0|0.2 |  
|90000| 0| 0.2| 1|0.8 |0.2 | 0| 0|0.8 |   
|100000| 0| 0| 1|0 |0 | 0| 0|0 |

```python
import numpy as np
import matplotlib.pyplot as plt 

# Khong gian nen 
X = np.array([5000, 6000, 9000, 10000, 20000, 50000, 60000, 90000, 100000])

# Ham thuoc
mu_A = np.array([0,0,0,0,0.3,1,0.8,0.2,0])
mu_B = np.array([0,0,0.8,1,0.3,0,0,0,0])

# chart
plt.figure()
plt.plot(X, mu_A, label = 'Tập mờ khách hàng chi tiêu lớn')
plt.plot(X, mu_B, label='Tập mờ khách hàng có lợi nhuận')
plt.title("Đồ thị tập mờ")
plt.xlabel("Khoảng chi tiêu")
plt.ylabel("Mức độ phụ thuộc μ(x)")
plt.legend()
plt.grid()
plt.show()
```
