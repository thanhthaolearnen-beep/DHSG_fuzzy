## Bài giải  

Theo đề bài ta có tập nền độ bền: X = {1,2,3,4}  

Độ bền kéo của mỗi loại thép:  

$\tilde{A}$ = $\frac{0.4}{1} + \frac{0.35}{2} + \frac{0.5}{3} + \frac{0.6}{4}$  

$\tilde{B}$ = $\frac{0.7}{1} + \frac{0.75}{2} + \frac{0.65}{3} + \frac{0.8}{4}$   

So sánh độ bền thép:  

|X|$\tilde{A}$|$\tilde{B}$|$\bar{A}$|$\bar{B}$|$\tilde{A}\cup\tilde{B}$|$\tilde{A}\cap\tilde{B}$|$\tilde{A} - \tilde{B}$|
|---|---|---|---|---|---|---|---|  
|1|0.4|0.7|0.6|0.3|0.7|0.4|0.3|  
|2|0.35|0.75|0.65|0.25|0.75|0.35|0.25|  
|3|0.5|0.65|0.5|0.35|0.65|0.5|0.35|  
|4|0.6|0.8|0.4|0.2|0.8|0.6|0.2|

```python
import numpy as np
import matplotlib.pyplot as plt

# Không gian nền
X = np.array([1,2,3,4])

# Hàm thuộc
mu_A = np.array([0.4,0.35,0.5,0.6])  
mu_B = np.array([0.7,0.75,0.65,0.8])  

plt.figure()
plt.plot(X, mu_A, label='Tập mờ A') 
plt.plot(X, mu_B, label='Tập mờ B') 
plt.title("Đồ thị mờ")
plt.xlabel("Độ bền thép")
plt.ylabel("Hai loại thép")
plt.legend()
plt.grid()
plt.show()
```
