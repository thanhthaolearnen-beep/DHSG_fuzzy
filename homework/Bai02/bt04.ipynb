```python
import numpy as np
import matplotlib.pyplot as plt

# 1: Không gian nền
X = np.array([10, 30, 100, 200, 500])

# 2. Hàm thuộc
mu_M = np.array([0.2,0.5,0.9,1,1])
mu_C = np.array([0, 0.3, 0.7, 0.9, 1])
mu_T = np.array([0,0,0.2,0.4,0.7]) 

# chart
plt.figure()
plt.plot(X, mu_M, label = 'Tập mờ Motorbike (M)')
plt.plot(X, mu_C, label='Tập mờ Car (C)')
plt.plot(X, mu_T, label='Tập mờ Truck (T)')
plt.title("Đồ thị tập mờ")
plt.xlabel("Công suất động cơ (HP)")
plt.ylabel("Mức độ phụ thuộc μ(x)")
plt.legend()
plt.grid()
plt.show()
```
