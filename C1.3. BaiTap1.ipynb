### Bài tập 1:
* Cho **X** là không gian nền về kích thước phòng học căn cứ vào số lượng bàn ghế đủ cho sinh viên ngồi học.

* Chẳng hạn, có 8 phòng học với khả năng chứa là một phần tử của tập **X** như sau:
 
**X = {30, 50, 75, 90, 100, 125, 150, 200}**.
  
* Xây dựng các tập:

  * **A** là tập mờ các phòng học rộng.
  * **B** là tập mờ các phòng có thể chứa 100 sinh viên.
  * **C** là tập các phòng học không chứa quá 100 sinh viên.

### Bài giải:  

Không gian nền X = {30, 50, 75, 90, 100, 125, 150, 200}  
Chọn 3 tập mờ:  

A = {$\frac{0}{30}, \frac{0}{50}, \frac{0}{75}, \frac{0}{90}, \frac{0.5}{100}, \frac{1}{125}, \frac{1}{150}, \frac{1}{200}$}  

B = {$\frac{0}{30}, \frac{0}{50}, \frac{0}{75}, \frac{0.6}{90}, \frac{1}{100}, \frac{1}{125}, \frac{1}{150}, \frac{1}{200}$}    

C = {$\frac{1}{30}, \frac{1}{50}, \frac{1}{75}, \frac{1}{90}, \frac{0}{100}, \frac{0}{125}, \frac{0}{150}, \frac{0}{200}$}    

Vẽ đồ thị:  

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Không gian nền  
X = np.array([30, 50, 75, 90, 100, 125, 150, 200])  

# Hàm thuộc  
mu_A = np.array([0,0,0,0,0.5,1,1,1])  
mu_B = np.array([0,0,0,0.6,1,1,1,1])  
mu_C = np.array([1,1,1,1,0,0,0,0])   

# Vẽ đồ thị
plt.figure(figsize=(8,5))
plt.plot(X, mu_A, marker='o', label='Tập mờ A')
plt.plot(X, mu_B, marker='o', label='Tập mờ B')
plt.plot(X, mu_C, marker='o', label='Tập mờ C')
plt.title("Đồ thị tập mờ")
plt.xlabel("Không gian nền X")
plt.ylabel("Mức độ phụ thuộc μ(x)")
plt.ylim(-0.05,1.05)
plt.legend()
plt.grid(True)
plt.show()
```

<span style="color:blue">Hàm thuộc liên tục</span>  

* Tập mờ A:  

$$\mu_A = \begin{cases}  
0,\quad x \in [30, 90) \\
\frac{x - 90}{20}, \quad x \in[90, 100) \\
\frac{125 - x}{50}, \quad x \in [100, 125) \\
1, \quad x \in [125, 200] \\
\end{cases}
$$  

* Tập mờ B:  

$$\mu_B = \begin{cases}  
0, \quad x \in [30, 75) \\
\frac{0.6x -45}{15}, \quad x \in [75, 90) \\
\frac{0.4x -30}{10}, \quad x \in [90, 100) \\
1, \quad x \in [100, 200] \\
\end{cases}  
$$

* Tập mờ C:  

$$\mu_C = \begin{cases}  
1, \quad x \le 90 \\
\frac{100 - x}{10}, \quad x \in [90, 100) \\
0, \quad x \in [100, 200] \\
\end{cases}  
$$
